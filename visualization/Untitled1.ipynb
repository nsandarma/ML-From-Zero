```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/drug200.csv')
```

```python
df.BP.values
```

```python
def conv(value):
    if value == 'HIGH':
        return 3
    elif value == 'NORMAL':
        return 2
    elif value == 'LOW':
        return 1
df['Cholesterol_num'] = df.Cholesterol.apply(conv)
df['BP_num'] = df.BP.apply(conv)
df
```
